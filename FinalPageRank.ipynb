```python
import networkx as nx
import numpy as np
from scipy.sparse import csc_matrix



def pageRank(G, s = .85, maxerr = .001):
    """
    Computes the pagerank for each of the n states.
    Used in webpage ranking and text summarization using unweighted
    or weighted transitions respectively.
    Args
    ----------
    G: matrix representing state transitions
       Gij can be a boolean or non negative real number representing the
       transition weight from state i to j.
    Kwargs
    ----------
    s: probability of following a transition. 1-s probability of teleporting
       to another state. Defaults to 0.85
    maxerr: if the sum of pageranks between iterations is bellow this we will
            have converged. Defaults to 0.001
    """
    n = G.shape[0]

    # transform G into markov matrix M
    M = csc_matrix(G,dtype=np.float64) #         
    
    rsums = np.array(M.sum(1))[:,0] #          0   
    
    ri, ci = M.nonzero() #index of valus not zero
    
    M.data /= rsums[ri] #m.data  

    # bool array of sink states
    sink = rsums==0

    # Compute pagerank r until we converg = np.zeros(n), np.ones(n)
    while np.sum(np.abs(r-ro)) > maxerr:
        ro = r.copy()
        # calculate each pagerank at a time
        for i in range(0,n):
            # inlinks of state i
            Ii = np.array(M[:,i].todense())[:,0]
            # account for sink states
            Si = sink / np.float64(n)
            # account for teleportation to state i
            Ti = np.ones(n) / np.float64(n)

            r[i] = ro.dot( Ii*s + Si*s + Ti*(1-s) )

    # return normalized pagerank
    return r/sum(r)


G = nx.DiGraph()
for link in the_links[:11]:
    G.add_node(link)


for link1 in the_links[:11]:
    for link2 in the_links[:11]:
       if ckeck_link_in_links(link1,link2):
           G.add_edges_from([(link1,link2)])


#G = nx.random_k_out_graph(n=8, k=2, alpha=0.75)
# G.add_node("https://www.bbc.com")
# G.add_node("https://edition.cnn.com/")
# G.add_node("https://edition.cnn.com/2022/12/22/europe/putin-uses-word-war-fighting-ukraine-russia-intl-hnk/index.html")
# G.add_edges_from([("https://www.bbc.com","https://edition.cnn.com/"),("https://edition.cnn.com/","https://edition.cnn.com/2022/12/22/europe/putin-uses-word-war-fighting-ukraine-russia-intl-hnk/index.html")])

# def draw_graph(G):
#     nx.draw_circular(G, node_size=400, with_labels=True)
    
# draw_graph(G)

M = nx.to_numpy_array(G)


rank_pages= pageRank(M,s=.86)
```

```python
path = "C:\\Users\\hosam\\Documents\\bigdata\\output2.txt"
method = 'r'


def decodeFormatFile(path, method):
    data = open(path, method, encoding="utf8")
    dataWordFormat = []
    uniqueWords = []
    for detailsWord in data.readlines():
        detailsWord = " ".join(detailsWord.split())
        dataWord = detailsWord.split(" ")
        uniqueWords.append(dataWord[0])
        
        dataWord[1] = dataWord[1].split(';')
        dataWord[1].pop()

        for i in range(0, len(dataWord[1])):
            dataWord[1][i] = dataWord[1][i].split(':')

        dataWordFormat.append(dataWord)
    data.close()
    
    return set(uniqueWords), dataWordFormat


words, word_with_urls_and_wordcount = decodeFormatFile(path, method)
```
